# 百度新聞高鐵時刻資料爬蟲	

## 匯入函式庫

```python
import requests #匯入 requests 函式庫 功能：向 http 請求
from bs4 import BeautifulSoup #從 bs4 函式庫 匯入 BeautifulSoup函式，來解析網頁。
```

## 索取網頁內容

```python
response = requests.get("https://news.baidu.com/")

#以 requests 函式庫內的 get()函式 來請求 http 的網頁內容。
```

## 網頁內容解析 

```python
soup = BeautifulSoup(response.text, 'html.parser')
#                      解析內容         解析模式

#透過 BeautifulSoup() 函式，來解析網頁內容。
```

```python
soup
```

## 網頁內容定位

```python
news = soup.select('#pane-news a')

#藉由 select()函式，來解析網頁的資料
```

```python
news
```

## 網頁內容爬取

```python
for item in news:
    print(item.text) #列印 網頁 文字內容
    print(item['href']) #列印 網頁 超連結內容
```
