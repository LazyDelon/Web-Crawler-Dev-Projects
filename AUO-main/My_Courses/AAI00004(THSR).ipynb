# 百度新聞高鐵時刻資料爬蟲

## 匯入函式庫

```python
import requests #匯入 requests 函式庫 功能：向 http 請求
```

## 索取網頁內容

```python
form = {
    'SearchType': 'S',
    'Lang': 'TW',
    'StartStation': 'TaiNan',
    'EndStation': 'TaiPei',
    'OutWardSearchDate': '2022/04/07',
    'OutWardSearchTime': '05:00',
    'ReturnSearchDate': '2022/04/07',
    'ReturnSearchTime': '05:00'
}
```

![title](Photo/Figure_2.jpg)

```python
res = requests.post("https://www.thsrc.com.tw/TimeTable/Search", data=form)

#以 requests 函式庫內的 post()函式，並以 form data 來請求 http 的網頁內容。
```

## 網頁內容解析

```python
import json #匯入 json 函式庫 功能：處理 json 型態資料
```

```python
data = json.loads(res.text)

#將爬取到的 res 透過 loads 來解析 json 格式
```

```python
data
```

```python
chooese = data['data']['DepartureTable']['TrainItem']

#從 data 中篩選出所需的資料欄位
```

```python
print("車次 \t", "出發 \t", "抵達 \t", "耗時 \t", "自由做 \t") #列印 標題

for item in chooese:
    print(item['TrainNumber'] + " \t", item['DepartureTime'] + " \t", item['DestinationTime'] + " \t", item['Duration'] + " \t", item['NonReservedCar'] + " \t")
```
